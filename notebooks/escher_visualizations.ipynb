# Escher pathway visualizations

## 1. Imports

```python
# Automatic module reloading
%load_ext autoreload
%autoreload 2

# Packages
import os
import sys
import escher
from escher import Builder
import cobra as cb
import pandas as pd

# Add parent directory 
sys.path.append(os.path.abspath('..'))

# Paths
model_xml = os.path.join("../models", f"iML1515_GEM.xml")
data_path = os.path.join("../notebooks", "data", "exp_validation_results")
```

## Standard Escher map - no flux info mapped

```python
escher.list_available_maps()
```

```python
# Model 
model = cb.io.read_sbml_model(model_xml)

# Builder
# Map could be changed to any of the above or create a new one
# iJO1366 is almost the same as iML1515
builder = Builder(map_name='iJO1366.Central metabolism', model=model)

# Configurations
builder .highlight_missing = True
builder.hide_secondary_metabolites = True
builder.reaction_label_style = {'font_size': 12}

# Display
builder
```

## Regular model

Reaction flux results from the iML1515 GEM constrained with kcat values using kinGEMs. No changes on model media.

```python
# Process results
regular_path = os.path.join(data_path,
                        "W3110_MD121_M9+Glu_none", 
                        "df_FBA_regular_model.csv")

df_regular = pd.read_csv(regular_path)
df_reactions = df_regular[df_regular['Variable'] == 'flux']
df_reactions = df_reactions[df_reactions['Value'] != 0] # drop 0 value fluxes
flux_dict_regular = pd.Series(df_reactions['Value'].values, index=df_reactions['Index']).to_dict()
print(flux_dict_regular)
```

```python
builder_regular = Builder(map_name='iJO1366.Central metabolism', model=model)

# Add fluxes to builder
builder_regular.reaction_data = flux_dict_regular

# Configurations
builder_regular .highlight_missing = True
builder_regular.hide_secondary_metabolites = True
builder_regular.reaction_label_style = {'font_size': 12}

# Display
builder_regular
```
